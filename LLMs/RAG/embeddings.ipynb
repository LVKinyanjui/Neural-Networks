```python
import os
import pprint
import google.generativeai as palm
```

```python
# Credentials
api_key = os.getenv("PALM_API_KEY")
palm.configure(api_key=api_key)
```

```python
# List Models
models = palm.list_models()
for model in models:
    print(model.name)
    # pprint.pprint(model)
```

```python
# Get models
model = palm.get_model(name='models/embedding-gecko-001')
```

```python
text = "I am getting text embeddings using the palm api"
```

```python
# Get embeddings
response = palm.generate_embeddings(text=text, model=model)
```
